```python
# For finding the missing values in the dataset :-

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
missing_value = ['n/a','-','?']
df = pd.read_csv("adult.csv",na_values=missing_value)
df.head()
```

```python
df.dropna(subset=['workclass','occupation'],axis=0,how='all',inplace=True)
df.head()
```

```python
# Removing 'fnlwgt' column from the dataset :-

df = df.drop(df.columns[[2]], axis=1)
df.head()
```

```python
# Numerical Features_
# Column Normalization with the help of 'MinMaxScaler' :-

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("adult.csv")
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
print(scaler.fit(df['age']))
print(scaler.data_max_)
print(scaler.data_min_)
print(scaler.transform(df['age']))
```

```python
scaler = MinMaxScaler()
print(scaler.fit(df['educational-num']))
print(scaler.data_max_)
print(scaler.data_min_)
print(scaler.transform(df['educational-num']))
```

```python
scaler = MinMaxScaler()
print(scaler.fit(df['capital-gain']))
print(scaler.data_max_)
print(scaler.data_min_)
print(scaler.transform(df['capital-gain']))
```

```python
scaler = MinMaxScaler()
print(scaler.fit(df['hours-per-week']))
print(scaler.data_max_)
print(scaler.data_min_)
print(scaler.transform(df['hours-per-week']))
```
